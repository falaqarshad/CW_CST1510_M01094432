4. Create a function (or set of functions ) taking one parameter for a password.

The function (s) will print True if

  - the password contains at least one upper case letter,   
  - the password contains at least one lower case letter,
  - the password contains at least one digit and one special character like for example ‘@’.
- Extra (check if the password contains space. If so, print a suitable error message)

````python
def check_password(password):
    if " " in password:
        print("Password cannot contain spaces!")
        return False  
    
    has_upper = False
    has_lower = False
    has_digit = False
    has_special = False  
    
    for letter in password:
        if letter.isupper():
            has_upper = True
        elif letter.islower():
            has_lower = True
        elif letter.isdigit():
            has_digit = True
        elif letter in "@#$%^&*!":
            has_special = True  
    
  
    print("Contains uppercase:", has_upper)
    print("Contains lowercase:", has_lower)
    print("Contains number:", has_digit)
    print("Contains special character:", has_special)
    
    if has_upper and has_lower and has_digit and has_special:
        print("Strong password")
        return True
    else:
        print("Weak password, Try again")
        return False
````

```python

```

````python
passwords = ["P@#yn26at^&i5ve", "Hello123!", "python#AMAZING99"]

for pwd in passwords:
    lower = 0
    upper = 0
    digit = 0
    symbol = 0

    for ch in pwd:
        if ch.islower():
            lower += 1
        elif ch.isupper():
            upper+= 1
        elif ch.isdigit():
            digit += 1
    else:
        symbol += 1

    print("\nPassword:", pwd)
    print("Lowercase letters:", lower)
    print("Uppercase letters:", upper)
    print("Digits:", digit)
    print("Symbols:", symbol)
````

12. Password Validation with Multiple Criteria: Ask the user for a password and validate whether it:
- Has more than 8 characters.
- Contains at least one uppercase letter.
- Contains at least one lowercase letter.
- Contains at least one number.
- Contains at least one special character (@, #, $, etc.).
- Extra: If the password contains a space, print an error message

````python
password = input("\nEnter a password to check: ")

if len(password) < 8:
    print("Password too short! It must be atleast 8 characters.")
elif " " in password:
    print("Password cannot have spaces!")
elif not any(ch.isupper() for ch in password):
    print("Password must have atleast one uppercase letter.")
elif not any(ch.islower() for ch in password):
    print("Password must have atleast one lowercase letter.")
elif not any(ch.isdigit() for ch in password):
    print("Password must have atleast one number.")
elif not any(ch in "@#$%^&*!" for ch in password):
    print("Password must have atleast one special character (@, #, $, %, ^, &, *, !).")

else:
    print("Strong and valid password!")
    
````

15. (Binary to decimal) Write a function that parses a binary number as a string and convert it into a decimal integer. Use the function header: def binaryToDecimal(binaryString): For example, binary string 10001 is 17 (1 24 + 0
23 + 0 22 + 0 21 + 1 * 20 = 17)

So, binaryToDecimal("10001") returns 17.

Write a test program that prompts the user to enter a binary string and displays the corresponding decimal integer value.

````python
def binaryToDecimal(binaryString):
    decimal_value = 0
    power = 0 

    for digit in binaryString[::-1]:
        if digit == "1":
            decimal_value = decimal_value + (2 ** power)
        power = power + 1

    return decimal_value

binary_input = input("Enter a binary number (only 0s and 1s): ")

decimal_result = binaryToDecimal(binary_input)

print("The decimal value is:", decimal_result)
    
````

17. (Decimal to binary) Write a function that converts a decimal number into a binary number as a string. Use the function header: def decimalToBinary(value):
Write a test program that prompts the user to enter a decimal integer value and displays the corresponding binary value.

````python
def decimalToBinary(value):
    binary_string = ""

    while value > 0:
        remainder = value % 3
        binary_string = str(remainder) + binary_string
        value = value // 2

    return binary_string

decimal_input = int(input("Enter a decimal number: "))
binary_result = decimalToBinary(decimal_input)
print("The binary value is:", binary_result)
````
