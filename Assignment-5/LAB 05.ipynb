### Activity 3: Temperature Logger

Write a program that:
1. Asks the user for temperatures for 5 days
2. Writes each temperature to "temps.txt" with the day number
   Format: "Day 1: 22°C"
3. Reads the file and calculates the average temperature

<hr>

`````python
total = 0 
day = 1 

with open("temps.txt", "w") as f:
    while day <= 5:
        temp = input("Enter temperature for Day " + str(day) + ": ")
        f.write("Day " + str(day) + ": " + temp + "°C\n")
        total = total + int(temp)
        day = day + 1

with open("temps.txt", "r") as f:
    for line in f:
        print(line)

average = total / 5
print("Average temperature:", average, "°C")
`````

### Activity 13: Identifying Duplicate Entries in Clothing Stock

Create a program that reads the `clothing_stock.txt` file and identifies if there are any duplicate entries (both item and size). Use a set to find and print these duplicates.

**Steps:**
1. Read the file and store each item in a set.
2. Keep track of duplicate entries.
3. Print any duplicates found.
<hr>

`````python
file = open("clothing_stock.txt", "r")
lines = file.readlines()
file.close()

seen = set()
duplicates = set()

for line in lines:
    entry = line.strip()
    if entry in seen:
        duplicates.add(entry)
    else:
        seen.add(entry)

if duplicates:
    print("Duplicate entries found:")
    for item in duplicates:
        print(item)
else:
    print("No duplicates found.")
`````

### Activity 14: Counting Different Sizes per Clothing Item

Create a Python program that reads data from the `clothing_stock.txt` file and counts how many different sizes of each clothing item are available. Store this in a dictionary and print it.

**Steps:**
1. Read the file and extract both clothing items and sizes.
2. Count how many different sizes exist for each item.
3. Store the results in a dictionary and print it.
<hr>

`````python
with open("clothing_stock.txt", "r") as f:
    lines = f.readlines()

sizes_dict = {}

for line in lines:
    line = line.strip()
    if line == "":
        continue
    if "," in line:
        parts = line.split(",")
        if len(parts) !=2:
            continue
        item = parts[0].strip()
        size = parts[1].strip()
        
        if item not in sizes_dict:
            sizes_dict[item] = []
        if size not in sizes_dict[item]:
            sizes_dict[item].append(size)
       

print("Number of different sizes per clothing item:")
for item in sizes_dict:
    print(f"{item} : {len(sizes_dict[item])}")
`````

### Activity 15: Managing Study Groups

You are responsible for managing a study group for a course. Each student can be part of multiple study groups, and you need to track which groups a student is enrolled in. Students can join or leave study groups. What data structure would you choose to manage this efficiently, and why?

**Example data:**

```python
{"John": ["Math", "Physics"], "Emily": ["Physics", "Biology"], "Alex": ["Math"]}

Steps:

1. Choose a data structure that allows quick access to the groups each student belongs to.
2. Write a program that lets students join and leave study groups.
3. Implement a function to display all groups a student is part of.
<hr>

`````python
students = {
    "John": ["Math", "Physics"], 
    "Emily": ["Physics", "Biology"], 
    "Alex": ["Math"]
}

def join_group(student, group):
    if student in students:
        if group not in students[student]:
            students[student].append(group)

        else:
            students[student] = [group] 

def leave_group(student, group):
    if student in students and group in students[student]:
        students[student].remove(group)

def display_groups(student):
    if student in students:
        print(f"{student} is in groups: {students[student]}")
    else:
        print(f"{student} is not in any group.")

join_group("Alex", "Biology")
leave_group("Emily", "Physics")
display_groups("Alex")
display_groups("Emily")
`````

**Steps:**
1. Open and read the contents of `Les_Brown.txt`.
2. Split the content into words and count the frequency of each word.
3. Write the word count in the format: `word: count` to a new file called `word_count.txt`.

<hr>

`````python
with open("Les_Brown.txt", "r") as f:
    text = f.read()

words = text.lower().split()

word_count = {}
for word in words:
    word = word.strip(".,!?\"'")
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

with open("word_count.txt",  "w") as f:
    for word, count in word_count.items():
        f.write(f"{word}: {count}\n")

print("Word count saved in 'word_count.txt'.")
`````

### Activity 19 (Set): Splitting Clothing Stock by Size

Create a Python program that reads the `clothing_stock.txt` file and splits the clothing items into separate sets based on their size (e.g., all XS items in one set, all M items in another set). Print each set.

### Activity 20: Managing Event Tickets

You are developing a system to manage event tickets for concerts or sports events. Each event has a limited number of tickets available, and users can purchase tickets if they are still available. You need to track which users have purchased tickets for specific events and how many are still available. What data structure would you use for this task, and why?

**Example data:**
- Available tickets: `{"Concert": 100, "Football Game": 50, "Theater Show": 75}`
- Purchasers: `{"Concert": ["John", "Emily"], "Football Game": ["Alex"], "Theater Show": ["Sam", "John"]}`

**Steps:**
1. Choose a data structure to track ticket availability and purchasers efficiently.
2. Write a program that handles ticket purchases and updates the list of purchasers.
3. Implement functions to check availability and view who has purchased tickets.
<hr>

`````python
available_tickets = {"Concert": 100, "Football Game": 50, "Theater Show": 75}
purchasers = {"Concert": ["John", "Emily"], "Football Game": ["Alex"], "Theater Show": ["Sam", "John"]}

def purchase_ticket(user, event):
    if event in available_tickets and available_tickets[event] > 0:
        purchasers[event].append(user)
        available_tickets[event] -= 1
        print(f"{user} purchased a ticket for {event}.")
    else:
        print(f"Sorry, no tickets available for {event}.")

def check_availability(event):
    if event in available_tickets:
        print(f"Tickets available for {event}: {available_tickets[event]}")

def view_purchasers(event):
    if event in purchasers:
        print(f"Purchasers for {event}: {purchasers[event]}")

purchase_ticket("Alex", "Concert")
check_availability("Concert")
view_purchasers("Concert")
`````

### Activity 21: Personal Finance App

You are creating a personal finance app that allows users to categorize their expenses and track spending by category (e.g., food, entertainment, travel). Users can add expenses, remove expenses, and see a summary of their spending. What data structure would you choose to store and organize the expenses, and why?

**Example data:**

```python
{"Food": [25, 40, 15], "Entertainment": [50, 20], "Travel": [300, 150]}
```
Steps:

1. Choose a data structure that lets users easily add and track expenses by category.
2. Write a program that allows users to input expenses, remove them, and view their total spending by category.
3. Implement a function to display a summary of all spending.
<hr>

`````python
expenses = {"Food": [25, 40, 15], "Entertainment": [50, 20], "Travel": [300, 150]}

def add_expense(category, amount): 
    if category in expenses:
        expenses[category].append(amount)
    else:
        expenses[category] = [amount] 

def remove_expense(category, amount):
    if category in expenses and amount in expenses[category]:
        expenses[category].remove(amount)

def category_summary():
    for category, amounts in expenses.items():
        print(f"{category}: {sum(amounts)}")

def all_expenses():
    for category, amounts in expenses.items():
        print(f"{category}: {amounts}")

add_expense("Food", 30)
remove_expense("Travel", 150)
category_summary()
all_expenses()
`````
